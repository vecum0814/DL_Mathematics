## Dataset for Regression

```python
import torch
import torch.nn as nn 
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

N, n_features = 8, 5 # y = w1*x1 + w2*x2 + w3*x3 + w4*x4 + w5*x5 + b

torch_weights = torch.tensor([1, 2, 3, 4, 5], dtype = torch.float32)
torch_bias = torch.tensor([10], dtype = torch.float32)

X = torch.randn(N, n_features)

Y = torch.sum(torch_weights * X, 1) + torch_bias

print('X(shape/dtype/data): {} / {} / {}\n'.format(X.shape, X.dtype, X.numpy()))
print('Y(shape/dtype/data): {} / {} / {}\n'.format(Y.shape, Y.dtype, Y.numpy()))
  
```

## Datset for Multi-class Classification

```python
import torch
import torch.nn as nn 
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')

N, n_features = 9, 2 # for x1, x2
n_class = 3

X = torch.zeros(0, n_features)
Y = torch.zeros(0, 1, dtype = torch.int32)

fig, axes = plt.subplots(figsize = (5, 5))

for class_idx in range(n_class):
    center = 15 * torch.clamp(torch.randn(2, 1), min = -1, max = 1) # coordinate values
    #axes.scatter(center[0], center[1])
    #print(center)
    x1 = center[0] + torch.randn(N, 1)
    x2 = center[1] + torch.randn(N, 1)    
    
    #axes.scatter(x1.numpy(), x2.numpy())
    x = torch.concat((x1, x2), axis = 1)
    y = class_idx * torch.ones(N, 1, dtype = torch.int32)
    #print(x.shape, y.shape)
    axes.scatter(x[:, 0].numpy(), x[:, 1].numpy(), alpha = 0.3)
    
    X = torch.concat((X, x), axis = 0)
    Y = torch.concat((Y, y), axis = 0)
    
    #print(X.shape, Y.shape)
    
print('X(shape/dtype/data): {} / {} / {}\n'.format(X.shape, X.dtype, X.numpy()))
print('Y(shape/dtype/data): {} / {} / {}\n'.format(Y.shape, Y.dtype, Y.numpy()))
```

## Dataset for Multi-class Classification with O-H-E

```python
import torch
import torch.nn as nn 
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')

N, n_features = 9, 2 # for x1, x2
n_class = 3

X = torch.zeros(0, n_features)
Y = torch.zeros(0, dtype = torch.int32)

fig, axes = plt.subplots(figsize = (5, 5))

for class_idx in range(n_class):
    center = 15 * torch.clamp(torch.randn(2, 1), min = -1, max = 1) # coordinate values
    #axes.scatter(center[0], center[1])
    #print(center)
    x1 = center[0] + torch.randn(N, 1)
    x2 = center[1] + torch.randn(N, 1)    
    
    #axes.scatter(x1.numpy(), x2.numpy())
    x = torch.concat((x1, x2), axis = 1)
    y = class_idx * torch.ones(N, dtype = torch.int32)
    #print(x.shape, y.shape)
    axes.scatter(x[:, 0].numpy(), x[:, 1].numpy(), alpha = 0.3)
    
    X = torch.concat((X, x), axis = 0)
    Y = torch.concat((Y, y), axis = 0)
    
    #print(X.shape, Y.shape)

Y = F.one_hot(Y.to(torch.int64), num_classes = 3)


print('X(shape/dtype/data): {} / {} / {}\n'.format(X.shape, X.dtype, X.numpy()))
print('Y(shape/dtype/data): {} / {} / {}\n'.format(Y.shape, Y.dtype, Y.numpy()))
```
