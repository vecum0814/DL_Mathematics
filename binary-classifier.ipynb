# The Graphs of Odds, Logit and Sigmoid

## The Graphs of Odds and Logit

```python
import torch
import torch.nn as nn 
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')

p_np = np.linspace(0.01, 0.99, 100)
p_torch = torch.linspace(0.01, 0.99, 100)

odds_np = p_np / (1 - p_np)
odds_torch = p_torch / (1 - p_torch)

logit_np = np.log(odds_np)
logit_torch = torch.log(odds_torch)

fig, axes = plt.subplots(2, 1, figsize = (15, 10),
                       sharex = True)
axes[0].plot(p_np, odds_np)
axes[1].plot(p_np, logit_np)

xticks = np.arange(0, 1.1, 0.2)
axes[0].tick_params(labelsize = 15)
axes[0].set_xticks(xticks)
axes[0].set_ylabel('Odds', fontsize = 20, color = 'darkblue')

axes[1].tick_params(labelsize = 15)
axes[1].set_xticks(xticks)
axes[1].set_xlabel('Probability', fontsize = 20, color = 'darkblue')
axes[1].set_ylabel('Logit', fontsize = 20, color = 'darkblue')
```

## The Graphs of Sigmoid

```python
import torch
import torch.nn as nn 
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt


x = torch.linspace(-10, 10, 100)

sigmoid = 1 / (1 + torch.exp(-x))

fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(x.numpy(), sigmoid.numpy())
```

# Logistic Regression Model

## Single-variate Logistic Regression Models

```python
import torch
import torch.nn as nn 
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')

X = torch.randn(100, 1)

class Dense(torch.nn.Module):
    def __init__(self):
        super(Dense, self).__init__()
        self.layer = nn.Linear(1, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        x = self.layer(x)
        x = self.sigmoid(x)
        return x
    
model = Dense()
Y = model(X)
print(X.size())
print(Y.size())


fig, ax = plt.subplots(figsize = (7, 7))
ax.scatter(X.detach().numpy().flatten(), Y.detach().numpy().flatten())
```
